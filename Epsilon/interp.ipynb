```python
import numpy as np
import pandas as pd

fidHB = pd.read_csv('epsilon.csv')
fidcol = pd.read_csv('collagen_mua.csv')
fidwater = pd.read_csv('water_mua.csv')
fidepi = pd.read_csv('mu_epi.csv')
fidmel = pd.read_csv('mel.csv')
```

```python
fidHB
fidcol
fidwater
fidepi
fidmel
```

```python
wl = [i for i in range(700, 900)]
wl
```

```python
mel = np.interp(wl, fidmel['wl'], fidmel['mel'])
mel
```

```python
dfmel = pd.DataFrame({'wl':wl, 'mel':mel})
dfmel
dfmel.to_csv('dfmel.csv')
```

```python
epi = np.interp(wl, fidepi['wl'], fidepi['epi'])
epi
```

```python
dfepi = pd.DataFrame({'wl':wl, 'epi':epi})
dfepi
dfepi.to_csv('dfepi.csv')
```

```python
hb = np.interp(wl, fidHB['wl'], fidHB['hb'])
hb
```

```python
hbo2 = np.interp(wl, fidHB['wl'], fidHB['hbo2'])
hbo2
```

```python
dfhb = pd.DataFrame({'wl':wl, 'hbo2':hbo2, 'hb':hb})
dfhb
dfhb.to_csv('dfhb.csv')
```

```python
col = np.interp(wl, fidcol['wl'], fidcol['col'])
col
```

```python
dfcol = pd.DataFrame({'wl':wl, 'col':col})
dfcol
dfcol.to_csv('dfcol.csv')
```

```python
dfwater = pd.DataFrame({'wl':wl, 'water':water})
dfwater
dfwater.to_csv('dfwater.csv')
```
